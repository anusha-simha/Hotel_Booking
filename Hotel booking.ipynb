```python
df['is_canceled'] = df['is_canceled'].astype(str)
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
new_labels = {'0': 'Not Canceled', '1': 'Canceled'}
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles, [new_labels[label] for label in labels], bbox_to_anchor=(1, 1))
plt.title('Reservation Status in Different Hotels', fontsize=20)
plt.xlabel('Hotel', fontsize=15)
plt.ylabel('Number of Reservations', fontsize=15)
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize= (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel') 
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x = 'month', hue='is_canceled', data=df, palette='bright')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation Status per month', size=20)
plt.xlabel('month')
plt.ylabel('Number of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load your dataset
df = pd.read_csv('hotel_booking.csv')

# Convert 'reservation_status_date' to datetime if not already
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Extract month from the date
df['month'] = df['reservation_status_date'].dt.month_name()

# Filter for canceled bookings
canceled_df = df[df['is_canceled'] == 1]

# Calculate average ADR per month for canceled bookings
adr_per_month = canceled_df.groupby('month')['adr'].mean().reset_index()

# Define the month order for plotting
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
adr_per_month['month'] = pd.Categorical(adr_per_month['month'], categories=month_order, ordered=True)
adr_per_month = adr_per_month.sort_values('month')

# Plot
plt.figure(figsize=(15, 8))
sns.barplot(x='month', y='adr', data=adr_per_month, palette='viridis')
plt.title('ADR per Month ', fontsize=30)
plt.xlabel('Month')
plt.ylabel('ADR')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('hotel_booking.csv')

# Filter for canceled reservations
cancelled_data = df[df['is_canceled'] == 1]

# Calculate top 10 countries by count
top_10_country = cancelled_data['country'].value_counts()[:10]

# Check if top_10_country has data
if top_10_country.empty:
    print("No data available for the pie chart.")
else:
    # Plot the pie chart
    plt.figure(figsize=(8, 8))
    plt.title('Top 10 Countries with Canceled Reservations')
    plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)
    plt.show()
```
